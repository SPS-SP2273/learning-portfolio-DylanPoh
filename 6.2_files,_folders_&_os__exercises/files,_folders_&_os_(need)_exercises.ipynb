# Task 1

```python
import numpy as np
import os
from glob import glob
import shutil

dict_month = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

filename = 'oct-08_singapore.txt'
date, city = filename.split('_')
month, day = date.split('-')
month = dict_month[month]
new_filename = f'{month}-{day}-{city}'
print(new_filename)
```

# Task 2

```python
def rename_my_file(old_file_name):
    try:
        date, city = old_file_name.split('_')
    except:
        print(f'{old_file_name} already renamed')
        return old_file_name
    month, day = date.split('-')
    try:
        month = dict_month[month]
    except:
        print(f'{old_file_name} already renamed')
    new_filename = f'{month}-{day}-{city}'
    return new_filename

rename_my_file('10-08-singapore.txt')
```

# Task 3

```python
filename_list = [
'oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
'jan-01_tokyo.txt']

for file in filename_list:
    new_filename = rename_my_file(file)
    print(f'{file} ----> {new_filename}')
```

# Task 4

```python
for path in glob('os-collab*/*.txt'):
    filename = os.path.split(path)[1]
    new_filename = rename_my_file(filename)
    new_path = os.path.join('os-collaboration-exercise-data', new_filename)
    if path != new_path:
        os.rename(path, new_path)
        print(f'Renaming {path} to {new_path}')
```

# Task 5

```python
for path in glob('os-collab*/*.txt'):
    filename = os.path.split(path)[1]
    country_txt = filename.split("-")[2]
    country = country_txt.split(".")[0]
    try:
        os.mkdir(country)
    except:
        print(f'{country} folder already exists')
```

# Task 6

```python
for path in glob('os-collab*/*cairo.txt'):
    filename = os.path.split(path)[1]
    path_to_destination = os.path.join("cairo", filename)
    shutil.copy(path, path_to_destination)
    print(f'Copied file to {path_to_destination}')
```
